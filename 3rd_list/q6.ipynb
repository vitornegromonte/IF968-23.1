```python
import math

# Função para calcular a distância euclidiana entre dois pontos
def distance(coord1, coord2):
    return math.sqrt((coord1[0] - coord2[0]) ** 2 + (coord1[1] - coord2[1]) ** 2)

# Inicialização das listas para cada zona da cidade
zonas_norte = []
zonas_sul = []
zonas_leste = []
zonas_oeste = []

# Leitura dos locais da cidade até a entrada "Fim"
while True:
    entrada = input()
    if entrada == "Fim":
        break

    zona, nome_local, x, y = entrada.split(" - ")
    x, y = int(x), int(y)

    # Mapeamento do local na matriz da respectiva zona
    if zona == "Norte":
        zonas_norte.append((nome_local, (x, y)))
    elif zona == "Sul":
        zonas_sul.append((nome_local, (x, y)))
    elif zona == "Leste":
        zonas_leste.append((nome_local, (x, y)))
    elif zona == "Oeste":
        zonas_oeste.append((nome_local, (x, y)))

# Impressão dos locais mapeados por zona
print("Cidade mapeada:")
print("Zona Norte")
if not zonas_norte:
    print("Nenhum local mapeado nessa zona!")
else:
    for nome_local, _ in zonas_norte:
        print(f"- {nome_local}")

print("Zona Sul")
if not zonas_sul:
    print("Nenhum local mapeado nessa zona!")
else:
    for nome_local, _ in zonas_sul:
        print(f"- {nome_local}")

print("Zona Leste")
if not zonas_leste:
    print("Nenhum local mapeado nessa zona!")
else:
    for nome_local, _ in zonas_leste:
        print(f"- {nome_local}")

print("Zona Oeste")
if not zonas_oeste:
    print("Nenhum local mapeado nessa zona!")
else:
    for nome_local, _ in zonas_oeste:
        print(f"- {nome_local}")

# Leitura da lista de esconderijos da Barbie
esconderijos_barbie = input().split(", ")

# Leitura da localização dos capangas e quantidade de quilômetros que podem percorrer
coordenada_x, coordenada_y, quilometros = map(int, input().split())
distancia_percorrida = 0

# Início da perseguição da Barbie enquanto houver gasolina suficiente
while quilometros > 0:
    # Encontrar o tipo de local mais próximo dos capangas a partir da lista de esconderijos da Barbie
    local_mais_proximo = None
    distancia_mais_proxima = float("inf")
    tipo_lugar = None

    for tipo_lugar in esconderijos_barbie:
        for nome_local, coordenadas in zonas_norte + zonas_sul + zonas_leste + zonas_oeste:
            dist = distance((coordenada_x, coordenada_y), coordenadas)
            if dist < distancia_mais_proxima:
                distancia_mais_proxima = dist
                local_mais_proximo = (nome_local, coordenadas)

    # Se o local mais próximo estiver mapeado, imprimir a mensagem
    if local_mais_proximo:
        print(f"Recebemos informações que a Barbie está no(a) {tipo_lugar} mais próximo... Vá atrás dela!")
    else:
        print(f"Nenhum(a) {tipo_lugar} encontrado na cidade! Que tipo de informação é essa!?")

    # Listar os locais mais próximos para os capangas guiarem-se
    print("Locais mais próximos:")
    for nome_local, coordenadas in zonas_norte + zonas_sul + zonas_leste + zonas_oeste:
        print(f"Coordenadas: ({coordenadas[0]}, {coordenadas[1]})")

    # Calcular a distância até o local mais próximo usando a fórmula da distância euclidiana
    distancia_mais_proxima = distance((coordenada_x, coordenada_y), local_mais_proximo[1])

    # Atualizar a localização dos capangas e o combustível gasto
    coordenada_x, coordenada_y = local_mais_proximo[1]
    quilometros -= distancia_mais_proxima
    distancia_percorrida += distancia_mais_proxima

    # Se o combustível acabar, imprimir a mensagem e encerrar a perseguição
    if quilometros <= 0:
        print("Não temos combustível suficiente para continuar a perseguição, infelizmente a Barbie conseguiu fugir!")
        break

    # Se o local encontrado for o último da lista de esconderijos da Barbie, imprimir a mensagem e encerrar a perseguição
    if tipo_lugar == esconderijos_barbie[-1]:
        print(f"Finalmente, depois de {distancia_percorrida:.2f}km percorridos, os capangas capturaram a Barbie no(a) {local_mais_proximo[0]} ({tipo_lugar}) - Coordenadas: ({coordenada_x}, {coordenada_y})")
        break

    # Caso contrário, atualizar a lista de esconderijos da Barbie, removendo o local encontrado, e prosseguir para o próximo passo da perseguição
    esconderijos_barbie.remove(tipo_lugar)
```
