```python
def calculo_posicao(astronauta, posicao_explosao, raio_impacto):
    coord_x = astronauta[0] - posicao_explosao[0] # x2 - x1
    coord_y = astronauta[1] - posicao_explosao[1] # y2 - y1
    
    distancia = ((coord_x) ** 2 + (coord_y) ** 2) ** 0.5 # aplicando a fórmula da distância euclidiana
    return distancia > raio_impacto # só retorna o valor caso a distância for maior que o raio do impacto, ou seja, se é um sobrevivente

def media_pontos_centrais(astronautas):
    if not astronautas: # Caso não tenham sobreviventes
        
        return None
    
    # iniciando as variáveis para as coordenadas do ponto central
    total_x = 0
    total_y = 0
    
    for coords in astronautas: # acessando as coordenadas da lista de astronautas
        total_x += coords[0]
        total_y += coords[1]
        
    # calculando a média 
    coord_central_x = total_x / len(astronautas)
    coord_central_y = total_y / len(astronautas)
    
    # declarando e retornando o ponto central
    ponto_central = [coord_central_x, coord_central_y]
    return ponto_central


def resgatar_astronautas(capsulas, posicao_explosao, raio_impacto):
    # iniciando o contador de sobreviventes 
    # iniciando uma lista vazia para receber os pontos centrais e depois calcular com a media_pontos_centrais
    num_sobreviventes = 0 
    pt_centrais = []
    
    for capsula in capsulas: # acessando uma capsula
        lista_sobreviventes = [] # criando uma lista vazia para receber as coordenadas dos sobreviventes
        
        for astronauta in capsula: # acessando as coordenadas do astronauta na capsula
            
            if calculo_posicao(astronauta, posicao_explosao, raio_impacto):
                num_sobreviventes += 1
                lista_sobreviventes.append(astronauta)
                
        if lista_sobreviventes: # Testa se a lista de sobreviventes não é vazia
            pt_central_capsula = media_pontos_centrais(lista_sobreviventes) # chama a função media_pontos_centrais e calcula o pt central
            pt_centrais.append(pt_central_capsula) # adicionando na lista vazia dos pontos centrais
        
    return [num_sobreviventes, pt_centrais] # retorna a contagem de sobreviventes e os pontos centrais

# Inputs: capsulas, posicao_explosao e raio_impacto
capsulas = eval(input())
posicao_explosao = eval(input())
raio_impacto = int(input())

# Imprime a função resgatar_astronautas
print(resgatar_astronautas(capsulas, posicao_explosao, raio_impacto))
```
