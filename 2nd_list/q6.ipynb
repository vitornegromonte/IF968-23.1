```python
# Inputs
nome_aluno = input()
nome_professor = input()

# Começando
print('E agora, só pela resenha:')
print(f'Melhor de 3 entre: {nome_aluno} e {nome_professor}!')
print('COMEÇOU!')

# Contador placar
vitoria_aluno = 0
vitoria_professor = 0

while vitoria_aluno < 2 and vitoria_professor < 2:
    pts_aluno = 0
    pts_professor = 0
    
    while pts_aluno < 11 or pts_professor < 11 or (pts_aluno - pts_professor < 2) or (pts_professor - pts_aluno < 2):
        
        num = int(input())
        
        if num % 2 == 0:
            pts_professor += 1
        else:
            pts_aluno += 1
        print(f'{pts_aluno} - {pts_professor}')
        
        
    if pts_aluno > pts_professor:
        vencedor = nome_aluno
        vitoria_aluno += 1
    else:
        vencedor = nome_professor
        vitoria_professor += 1
    print(f'{vencedor} ganhou esta partida!')
    print(f'Placar: {nome_aluno} [{vitoria_aluno}-{vitoria_aluno}] {nome_professor}')

print('FIM DA SÉRIE!')

if vitoria_aluno > vitoria_professor:
    print(f'{nome_aluno} ganhou a série! Puro talento!')
else:
    print(f'{nome_professor} ganhou a série! A experiência ganhou!')
```
