```python
# Inputs
## Time 1
time1_nome1 = input()
time1_nome2 = input()

## Time 2
time2_nome1 = input()
time2_nome2 = input()

## Número de partidas
qt_partidas = int(input())

# Ordem placar: [time1, time2]
vit_time1 = 0
vit_time2 = 0
total_pts_time1 = 0
total_pts_time2 = 0

# Começando o programa:
print(f'VAMO VER QUEM VAI COMER GRAMA! TEREMOS {qt_partidas} PARTIDAS ENTRE:')
print(f'{time1_nome1} e {time1_nome2} X {time2_nome1} e {time2_nome2}')

for i in range(qt_partidas):
    
    # Ordem: [time1, time2]
    pts_time1 = int(input())
    pts_time2 = int(input())
    # Atualizando os pontos de cada time:
    total_pts_time1 += pts_time1
    total_pts_time2 += pts_time2
    
    # Desistências
    ## Time 1 pontua e Time 2 não pontua
    if pts_time1 > 0 and pts_time2 == 0:
        print('JOGO ENCERRADO POR DESISTÊNCIA DA EQUIPE 2, QUE VERGONHA')
        vit_time1 += 1
        
    ## Time 1 não pontua e Time 2 pontua
    elif pts_time1 == 0 and pts_time2 > 0:
        print('JOGO ENCERRADO POR DESISTÊNCIA DA EQUIPE 1, QUE VERGONHA')
        vit_time2 += 1
        
    else:
        
        if pts_time1 > pts_time2:
            vit_time1 += 1
            
        elif pts_time1 < pts_time2:
            vit_time2 += 1
            
        elif pts_time1 == pts_time2:
            vit_time1 += 1
            
# Checando o número de partidas, definindo as métricas gerais e declarando o vencedor
if pts_time1 != 0 and pts_time2 != 0 and vit_time1 + vit_time2 == qt_partidas:
    pontos_totais_time1 = total_pts_time1 * (vit_time1/qt_partidas)
    pontos_totais_time2 = total_pts_time2 * (vit_time2/qt_partidas)
    
    total_disputa = total_pts_time1 + total_pts_time2
    
    ## Declarando o vencedor dos confrontos:
    # Time 1 wins
    if pontos_totais_time1 > pontos_totais_time2:
        print(f'CARAMBA! Tivemos um total de {total_disputa} bolas ao chão ao longo dessa disputa.')
        print(f'{time1_nome1} e {time1_nome2} são os grandes vencedores!')
        print(f'Mataram {total_pts_time1} bolas, ganhando {vit_time1} partidas')
    
    # Time 2 wins
    elif pontos_totais_time1 < pontos_totais_time2:
        print(f'CARAMBA! Tivemos um total de {total_disputa} bolas ao chão ao longo dessa disputa.')
        print(f'{time2_nome1} e {time2_nome2} são os grandes vencedores!')
        print(f'Mataram {total_pts_time2} bolas, ganhando {vit_time2} partidas')
```
