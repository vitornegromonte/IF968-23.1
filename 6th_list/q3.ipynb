```python
def adicionar_suspeito(suspeitos, nome):
    suspeitos[nome] = {}

def atualizar_suspeito(suspeitos, nome, caracteristicas):
    if nome in suspeitos:
        for caracteristica in caracteristicas:
            c, v = caracteristica.split(': ')
            suspeitos[nome][c] = v

def remover_suspeito(suspeitos, nome):
    if nome in suspeitos:
        del suspeitos[nome]

def julgar_suspeito(suspeitos, nome):
    
    if nome in suspeitos:
        if 'Biceps' in suspeitos[nome] and int(suspeitos[nome]['Biceps'][:-2]) > 45:
            biceps = True
        else:
            biceps = False

        if 'Treino' in suspeitos[nome]:
            tempo = suspeitos[nome]['Treino'].split()
            if tempo[1] == 'segundos':
                treino = int(tempo[0])
            elif tempo[1] == 'minutos':
                treino = int(tempo[0]) * 60
            elif tempo[1] == 'horas':
                treino = int(tempo[0]) * 3600
            else:
                treino = 0
        else:
            treino = 0

        if 'Frequencia' in suspeitos[nome]:
            frequencia = int(suspeitos[nome]['Frequencia'].split()[0])
        else:
            frequencia = 0

        if 'BF' in suspeitos[nome]:
            bf = int(suspeitos[nome]['BF'][:-1])
        else:
            bf = 0

        if 'Suor' in suspeitos[nome]:
            suor = suspeitos[nome]['Suor'] != 'Seco'
        else:
            suor = False

        if biceps and treino >= 1800 and frequencia >= 4 and bf < 10 and not suor:
            return 'Natural'
        else:
            return 'FAKE NATTY! USOU O SUCO!'
    else:
        return 'Quem é esse crazy man? Não tá aqui na database'

def natty_meter(suspeitos):
    total_suspeitos = len(suspeitos)
    natty_count = 0

    for nome in suspeitos:
        if julgar_suspeito(suspeitos, nome) == 'Natural':
            natty_count += 1

    if natty_count == 0:
        return 'Oh yeah, vivemos em uma sociedade sem suco, um great day'
    else:
        porcentagem = (natty_count / total_suspeitos) * 100
        return f'NOOO! {int(porcentagem)} USARAM O SUCO'

def main():
    suspeitos = {}

    while True:
        comando = input()
        
        if comando == 'FIM':
            break
        elif comando == 'Adicionar suspeito':
            nome = input()
            adicionar_suspeito(suspeitos, nome)
            print(f'Novo suspeito: {nome}')
            
        elif comando == 'Atualizar suspeito':
            nome = input().split('-> ')[0]
            caracteristicas = input().split(', ')
            atualizar_suspeito(suspeitos, nome, caracteristicas)
            
        elif comando == 'Remover suspeito':
            nome = input()
            remover_suspeito(suspeitos, nome)
            print(f'{nome} removido da lista de suspeitos, está limpo')
            
        elif comando == 'Julgamento':
            nome = input()
            resultado = julgar_suspeito(suspeitos, nome)
            print(f'Eu já tenho o meu veredito, e o meu veredito, {nome}:')
            print(resultado)
            
        elif comando == 'NattyMeter':
            print('NattyMeter, ON!')
            print(natty_meter(suspeitos))

if __name__ == "__main__":
    main()
```
